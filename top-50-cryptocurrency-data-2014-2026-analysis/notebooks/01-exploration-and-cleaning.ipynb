```python
import kagglehub
import numpy as np
import os
import polars as pl
import shutil
```

```python
try:
    print("Downloading dataset from KaggleHub...")
    path: str = kagglehub.dataset_download("kaushalnandania/crypto-data-2014-2026")
    print("Dataset successfully loaded.")
except Exception as e:
    print(f"Dataset unsuccessfully loaded. Error: {e}")

dest_path:str = "../data/raw"

try:
    print("Moving dataset from path to ../data/raw")
    for filename in os.listdir(path):
        shutil.move(os.path.join(path, filename), os.path.join(dest_path, filename))
    print("Dataset successfully moved to ../data/raw")
except Exception as e:
    print(f"Dataset unsuccessfully moved. Error {e}")

del path, dest_path
```
